**Loading Dataset1**

```
# Importing the libraries
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from matplotlib import colors
%matplotlib inline
from sklearn.svm import SVC
from sklearn import metrics


# Importing the dataset
col_names=['x1', 'x2', 't']
dataset1 = pd.read_csv('dataset1.csv', names=col_names, header=None)
X1= dataset1[['x1','x2']]
Y1= dataset1[['t']]

```

```
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X1_train, X1_test,Y1_train, Y1_test= train_test_split(X1,Y1, test_size=0.25, random_state=0)

```

**Loading Dataset 2**

```
# Importing the dataset
col_names=['x1', 'x2', 't']
dataset2 = pd.read_csv('dataset2.csv', names=col_names, header=None)
X2= dataset2[['x1','x2']]
Y2= dataset2[['t']]
```

```
# Splitting the dataset into the Training set and Test set
X2_train, X2_test,Y2_train, Y2_test= train_test_split(X2,Y2, test_size=0.25, random_state=0)
```

**Loading Dataset 3**

```
# Importing the dataset
col_names=['x1', 'x2', 't']
dataset3 = pd.read_csv('dataset3.csv', names=col_names, header=None)
X3= dataset3[['x1','x2']]
Y3= dataset3[['t']]
```

```
# Splitting the dataset into the Training set and Test set
X3_train, X3_test,Y3_train, Y3_test= train_test_split(X3,Y3, test_size=0.25, random_state=0)
```

# **Experiment(1): -** SVC classifier on Dataset1

```
# Fitting SVM to the Training set
clf1_svc= SVC(kernel="linear", gamma=0.025)
clf1_svc.fit(X1_train, Y1_train )

# Predicting the Test set results
Y1_pred_svc= clf1_svc.predict(X1_test)

# Measuring Accuracy
accuracy1_svc= clf1_svc.score(X1_test, Y1_test)
print("Accuracy = ", accuracy1_svc)
```

#**Ans 1)**
The model predicts with 100% accuracy on test set

#**Experiment (2): -** Plotting  decision boundary for experiment (1)

```
x1lim = (X1.iloc[:,0].min(), X1.iloc[:,0].max())
y1lim = (X1.iloc[:,1].min(), X1.iloc[:,1].max())
xx1, yy1 = np.meshgrid(np.linspace(x1lim[0], x1lim[1], 71),
                     np.linspace(y1lim[0], y1lim[1], 81))

Z1_svc = clf1_svc.predict(np.c_[xx1.ravel(), yy1.ravel()])
Z1_svc = Z1_svc.reshape(xx1.shape)
#Z1_svc = Z1_svc[:, 1].reshape(xx1.shape)


# Plot the results
fig = plt.figure(figsize=(5, 3.75))
ax1_svc = fig.add_subplot(111)
ax1_svc.scatter(X1.iloc[:, 0], X1.iloc[:, 1], c=['#FFE888' if sh==1 else'#FF9999' for sh in dataset1['t']], cmap=plt.cm.binary, zorder=2)

ax1_svc.contour(xx1, yy1, Z1_svc, [0.5], colors='k')

ax1_svc.set_xlim(x1lim)
ax1_svc.set_ylim(y1lim)

ax1_svc.set_xlabel('$x1$')
ax1_svc.set_ylabel('$x2$')

plt.show()
```

Decision boundary has been plotted

#**Experiment(7): -** Changing Hyperparameters

```
from sklearn.model_selection import GridSearchCV 

# defining parameter range 
param_grid = {'C': [0.001, 0.01, 0.1, 1.0], 
			'degree': [3,5,7], 
			'kernel': ['poly'],
      'coef0' : [0.01, 0.1, 1.0]} 

grid = GridSearchCV(SVC(), param_grid, refit = True) 

# fitting the model for grid search 
grid.fit(X3_train, Y3_train) 

# print best parameter after tuning 
print("Best Parameters are: -", grid.best_params_) 


# Predicting the Test set results
Y3_pred_grid= grid.predict(X3_test)

# Measuring Accuracy
accuracy3_grid= grid.score(X3_test, Y3_test)
print("Accuracy = ", accuracy3_grid)

# Making the Confusion Matrix
from sklearn import metrics
cnf_matrix3_grid = metrics.confusion_matrix(Y3_test, Y3_pred_grid)
print("Confusion Matrix")
print(cnf_matrix3_grid)


Z3_grid = grid.predict(np.c_[xx3.ravel(), yy3.ravel()])
Z3_grid = Z3_grid.reshape(xx3.shape)


# Plot the results
fig = plt.figure(figsize=(5, 3.75))
ax = fig.add_subplot(111)
ax.scatter(X3.iloc[:, 0], X3.iloc[:, 1], c=['#FFE888' if sh==1 else'#FF9999' for sh in dataset3['t']], cmap=plt.cm.binary, zorder=2)

ax.contour(xx3, yy3, Z3_grid, [0.5], colors='k')

ax.set_xlim(x3lim)
ax.set_ylim(y3lim)

ax.set_xlabel('$x1$')
ax.set_ylabel('$x2$')

plt.show()
```
